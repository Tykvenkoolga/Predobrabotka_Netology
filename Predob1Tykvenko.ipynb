### а)  Заполнить пропуски в столбце total_bedrooms средними значениями по категориям столбца ocean_proximity

```python
houses['total_bedrooms']= houses.groupby("ocean_proximity")['total_bedrooms'].transform(lambda x: x.fillna(x.mean()))
houses.head()
```
